<h3>A Breast Cancer Classification</h3>

```python
#importing libs
```

```python
import sklearn as sk
from sklearn.datasets import load_breast_cancer
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

<h5>--- now i'll be import and loading the data sets from Sklearn into a "Data" varialiable.</h5>

```python
#--- loading into variable Data 
data = load_breast_cancer()
```

<h4>-- Organizing Data/ loading it.<h4>

```python
label_names = data['target_names']
labels = data['target']
feature_names = data['feature_names']
features = data['data']
```

<h4> taking a look</h4>

```python
# observing
print(label_names)
print(labels[0])
print(feature_names[0])
print(features[0])
```

<h4>Visualization via Matplotlib<h4>

```python
#creating a plot
plt.figure(figsize=(8, 6))

# classifying them
for i, label_name in enumerate(label_names):
    plt.scatter(
        features[labels == i, 0],  # radius mean
        features[labels == i, 1],  # texture mean
        label=label_name
    )

# labels and title
plt.xlabel(feature_names[0])  # radius
plt.ylabel(feature_names[1])  # texture
plt.title("Breast Cancer Dataset: Mean Radius vs Mean Texture")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# spliting the data

train, test, train_labels, test_labels = train_test_split(features, labels, test_size=0.33,random_state=42)
```

<h2>Building Model based on GassianNB</h2>

```python
#initialiazing to gnb
gnb = GaussianNB()

#now i will be training it on the training data

model = gnb.fit(train, train_labels)
```

<h2>Making Predictions</h2>
<h5>we train the model, we can then use the trained model to make
predictions on the test set</h5>

```python
prediction = gnb.predict(test)
print(prediction)
```

<h5>the prediction returns an array of 1's and 0's which basically represents the predicted values related to the tumor class. meaning it being malignant or benign</h5>

<h2>Evaluation</h2>
<h5>i will be using the array of true class labels, helping evaluate the accuracy of the model's predeicted values by comparing the two arrays. i will be using the function 'accuracy_score()' from 'sklearn' library. </h5>

```python
#storing results in variable evaluation.
evaluation = accuracy_score(test_labels, prediction)

print(evaluation)
```

<h5>As you see in the output, the NB classifier is 94.15% accurate. This
means that 94.15 percent of the time the classifier is able to make the
correct prediction as to whether or not the tumor is malignant or benign
</h5>
